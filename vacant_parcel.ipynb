```python
import os
import geopandas as gpd
import pandas as pd

vacant_ids = set(vacant_parcel['parcel_id'])

# 2. 设置输入输出路径
input_folder = "/Users/houpuli/Redlining Lab Dropbox/HOUPU LI/ceshi/tidyparcel_feeds"
output_folder = "/Users/houpuli/Redlining Lab Dropbox/HOUPU LI/ceshi/vacant_parcel_output"
os.makedirs(output_folder, exist_ok=True)

# 3. 遍历所有包含 "parcel" 的文件
for filename in os.listdir(input_folder):
    if "parcel" in filename:
        input_path = os.path.join(input_folder, filename)

        try:
            # 读取文件
            gdf = gpd.read_file(input_path)

            # 筛选出 parcel_id 存在于 vacant_parcel 的记录
            filtered = gdf[gdf['parcel_id'].isin(vacant_ids)]

            # 如果有匹配结果就导出
            if not filtered.empty:
                # 构建新文件名：替换原扩展名为 .vacant_parcel
                base_name, ext = os.path.splitext(filename)
                output_filename = base_name + "_vacant" + ext
                output_path = os.path.join(output_folder, output_filename)

                filtered.to_file(output_path, driver="GeoJSON")
                print(f"✅ 已保存: {output_filename} ({len(filtered)} 条记录)")
            else:
                print(f"⚠️ 无匹配记录: {filename}")

        except Exception as e:
            print(f"❌ 错误文件: {filename}，原因: {e}")
```

```python
import os
import geopandas as gpd
import pandas as pd

# 设置路径
input_folder = "/Users/houpuli/Redlining Lab Dropbox/HOUPU LI/ceshi/vacant_parcel_output"

# 储存城市名与parcel数量
city_list = []
parcel_count_list = []

# 遍历所有.parcel文件
for filename in os.listdir(input_folder):
    if filename.endswith(".parcel"):
        city_name = filename.replace("_vacant.parcel", "")
        file_path = os.path.join(input_folder, filename)
        
        try:
            gdf = gpd.read_file(file_path)
            unique_count = len(gdf['parcel_id'].unique())
        except Exception as e:
            print(f"❌ Error reading {filename}: {e}")
            unique_count = None
        
        city_list.append(city_name)
        parcel_count_list.append(unique_count)

# 创建 DataFrame
result_df = pd.DataFrame({
    "city": city_list,
    "vacant_parcel_count": parcel_count_list
})

# 导出为 Excel 文件
output_excel_path = "/Users/houpuli/Redlining Lab Dropbox/HOUPU LI/ceshi/vacant_parcel_count_summary.xlsx"
result_df.to_excel(output_excel_path, index=False)

print("✅ Excel 导出完成：", output_excel_path)
```

```python
tidyparcel_feeds = gpd.read_file(r"/Users/houpuli/Dropbox/ceshi/tidyparcel_feeds.geojson")
# 1. Filter out the centroid rows
lines = tidyparcel_feeds[tidyparcel_feeds['side'] != 'centroid']

# 2. Define a helper to turn a group of lines into a (multi)polygon
def lines_to_polygon(geom_series):
    merged = unary_union(geom_series)
    polys = list(polygonize(merged))
    if len(polys) == 1:
        return polys[0]
    else:
        return MultiPolygon(polys)

# 3. Apply per parcel_id
records = []
for pid, grp in lines.groupby('parcel_id'):
    poly = lines_to_polygon(grp.geometry)
    records.append({'parcel_id': pid, 'geometry': poly})

# 4. Build the new GeoDataFrame
tidyparcel_poly = gpd.GeoDataFrame(records,
                                   crs=tidyparcel_feeds.crs,
                                   geometry='geometry')
tidyparcel_poly
```

```python
building_overture = gpd.read_file(r"/Users/houpuli/Dropbox/ceshi/building_overture/OneDrive_1_5-5-2025.zip")
building_overture['centrid'] = building_overture['geometry'].centroid
building_overture = gpd.GeoDataFrame(building_overture, geometry='centrid')
building_overture
```
